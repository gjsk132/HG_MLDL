```python
# plotting a 3D scatter plot
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

# 비어있는 데이터 생성
num_points = 10
X_train = {
    'width': np.random.rand(num_points) * 10,  # 랜덤한 가로 길이
    'height': np.random.rand(num_points) * 5,   # 랜덤한 세로 길이
    'color_score': np.random.rand(num_points) * 20  # 랜덤한 색상 점수
}

y_train = np.random.rand(num_points) * 10  # 색상 값 배열 생성

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train['width'], X_train['height'], X_train['color_score'], c=y_train, marker='o', s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```
